# 연속형 컬럼 이상치 탐지

```python
numeric_cols = ['총 생성 배아 수', '미세주입된 난자 수', '미세주입에서 생성된 배아 수', '저장된 배아 수', '수집된 신선 난자 수', '혼합된 난자 수', '미세주입된 난자 수', "미세주입 후 저장된 배아 수", "해동 난자 수", "저장된 신선 난자 수",
                "파트너 정자와 혼합된 난자 수", "기증자 정자와 혼합된 난자 수", "배아 이식 경과일"]

def out_iqr(df):
  Q1 = df.quantile(0.25) #하위25퍼
  Q3 = df.quantile(0.75) #상위25퍼
  IQR = Q3-Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  return lower_bound, upper_bound

for col in numeric_cols:
  lower, upper = out_iqr(train_copy[train_copy[col] != -1][col])
  train_copy.loc[(train_copy[col] < lower), col] = lower
  train_copy.loc[(train_copy[col] >= upper), col] = upper
  test_copy.loc[(test_copy[col] < lower), col] = lower
  test_copy.loc[(test_copy[col] >= upper), col] = upper

for col in numeric_cols:
  print(f"============={col}===============")
  print(f"결측 : {(train_copy[col].isnull().sum() / 256351 * 100)} %")
  values = train_copy[col].value_counts()
  for t, count in values.items():
        print(f"{t}: {count/256351*100}%")
```
