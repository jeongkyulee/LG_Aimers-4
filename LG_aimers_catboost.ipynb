```python
# 1️⃣ 라이브러리 불러오기
import pandas as pd
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

# 2️⃣ 데이터 로드
train = pd.read_csv('', sep=',')
test = pd.read_csv('', sep=',')

# 3️⃣ 타겟 컬럼 설정
target = '임신 성공 여부'

# 4️⃣ 학습 & 검증 데이터 분할 (80% 학습, 20% 검증)
train_data, val_data = train_test_split(train, test_size=0.2, random_state=42, stratify=train[target])

# 5️⃣ AutoGluon 모델 학습
predictor = TabularPredictor(label=target, eval_metric='roc_auc').fit(train_data=train_data, time_limit=600)  # 10분 내 최적 모델 탐색

# 6️⃣ 검증 데이터 평가
y_val = val_data[target]
y_val_pred_proba = predictor.predict_proba(val_data)  # 확률값 전체 가져오기

# 🔹 예측 확률 데이터 형식 확인 후 처리
if isinstance(y_val_pred_proba, pd.DataFrame):
    y_val_pred = y_val_pred_proba.iloc[:, 1]  # 데이터프레임이면 iloc 사용
else:
    y_val_pred = y_val_pred_proba[:, 1]  # numpy 배열이면 직접 사용

# 🔹 이진 예측값 (0 또는 1) 변환
y_val_binary = (y_val_pred >= 0.5).astype(int)

# 7️⃣ 모델 평가 지표 계산
accuracy = accuracy_score(y_val, y_val_binary)
f1 = f1_score(y_val, y_val_binary)
auc = roc_auc_score(y_val, y_val_pred)

# 8️⃣ 평가 결과 출력
print(f"🔹 Accuracy: {accuracy:.4f}")
print(f"🔹 F1-score: {f1:.4f}")
print(f"🔹 AUC: {auc:.4f}")

# 9️⃣ 테스트 데이터 예측
y_test_pred_proba = predictor.predict_proba(test)  # 테스트 데이터 확률 예측

# 🔹 예측 확률 데이터 형식 확인 후 처리
if isinstance(y_test_pred_proba, pd.DataFrame):
    y_test_pred = y_test_pred_proba.iloc[:, 1]  # 데이터프레임이면 iloc 사용
else:
    y_test_pred = y_test_pred_proba[:, 1]  # numpy 배열이면 직접 사용

# 🔟 제출 파일 생성
submission = pd.read_csv('')
submission['임신 성공 여부'] = y_test_pred
submission.to_csv('', index=False)

print("🚀 AutoGluon 최적 모델 예측 완료! 제출 파일 생성됨: submission_autogluon.csv")

# 11️⃣ Colab에서 제출 파일 다운로드
from google.colab import files
files.download('')


```

```python
# 1️⃣ 라이브러리 불러오기
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

# 2️⃣ H2O 초기화
h2o.init()

# 3️⃣ 데이터 로드
train = pd.read_csv('', sep=',')
test = pd.read_csv('', sep=',')

# 4️⃣ 타겟 컬럼 설정
target = '임신 성공 여부'

# 5️⃣ 학습 & 검증 데이터 분할 (80% 학습, 20% 검증)
train_data, val_data = train_test_split(train, test_size=0.2, random_state=42, stratify=train[target])

# 6️⃣ Pandas → H2OFrame 변환
train_h2o = h2o.H2OFrame(train_data)
val_h2o = h2o.H2OFrame(val_data)
test_h2o = h2o.H2OFrame(test)

# 7️⃣ 타겟 변수 설정 (범주형 변수로 변환)
train_h2o[target] = train_h2o[target].asfactor()
val_h2o[target] = val_h2o[target].asfactor()

# 8️⃣ H2O AutoML 모델 학습 (최대 실행 시간: 10분)
aml = H2OAutoML(max_runtime_secs=600, seed=42)
aml.train(y=target, training_frame=train_h2o)

# 9️⃣ 최적 모델 선택
best_model = aml.leader
print(f"✅ 선택된 최적 모델: {best_model.algo}")

# 🔟 검증 데이터 평가
y_val = val_data[target]
y_val_pred_proba = best_model.predict(val_h2o).as_data_frame()['p1']  # 클래스 1의 확률값

# 🔹 이진 예측값 (0 또는 1) 변환
y_val_binary = (y_val_pred_proba >= 0.5).astype(int)

# 1️⃣1️⃣ 모델 평가 지표 계산
accuracy = accuracy_score(y_val, y_val_binary)
f1 = f1_score(y_val, y_val_binary)
auc = roc_auc_score(y_val, y_val_pred_proba)

# 1️⃣2️⃣ 평가 결과 출력
print(f"🔹 Accuracy: {accuracy:.4f}")
print(f"🔹 F1-score: {f1:.4f}")
print(f"🔹 AUC: {auc:.4f}")

# 1️⃣3️⃣ 테스트 데이터 예측
y_test_pred_proba = best_model.predict(test_h2o).as_data_frame()['p1']  # 클래스 1의 확률값
y_test_pred = (y_test_pred_proba >= 0.5).astype(int)

# 1️⃣4️⃣ 제출 파일 생성
submission = pd.read_csv('')
submission['임신 성공 여부'] = y_test_pred
submission.to_csv('', index=False)

print("🚀 H2O AutoML 최적 모델 예측 완료! 제출 파일 생성됨: submission_h2o_automl.csv")

# 1️⃣5️⃣ Colab에서 제출 파일 다운로드
from google.colab import files
files.download('')

# 1️⃣6️⃣ H2O 세션 종료
h2o.shutdown(prompt=False)
```
